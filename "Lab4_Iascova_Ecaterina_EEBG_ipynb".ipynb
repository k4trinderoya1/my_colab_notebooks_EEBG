```python
import pandas as pd

url = 'https://data.cityofnewyork.us/api/views/uvpi-gqnh/rows.csv?accessType=DOWNLOAD'
lab1_df = pd.read_csv(url)
print(lab1_df.head())

columns_of_interest = ['spc_latin', 'spc_common', 'status', 'health', 'tree_dbh', 'stump_diam']
lab1_df_reduced = lab1_df[columns_of_interest]
print(lab1_df_reduced.head())

nullValue_counts = lab1_df_reduced.isnull().sum()
print(nullValue_counts)
NaNsOnly = lab1_df_reduced.loc[(lab1_df_reduced['spc_latin'].isnull()) | (lab1_df_reduced['spc_common'].isnull()) | (lab1_df_reduced['health'].isnull())]
print(NaNsOnly.head())
print(pd.value_counts(NaNsOnly['status']))

zeros_count = lab1_df_reduced.loc[(lab1_df_reduced['tree_dbh'] == 0) & (lab1_df_reduced['stump_diam'] == 0)]
print(len(zeros_count))
lab1_df_reduced_zerosFiltered = lab1_df_reduced.loc[(lab1_df_reduced['tree_dbh'] != 0) | (lab1_df_reduced['stump_diam'] != 0)]
print(len(lab1_df_reduced_zerosFiltered))
group_columns = ['status', 'spc_latin', 'spc_common', 'health']
lab1_df_reduced_zerosFiltered_grouped = lab1_df_reduced_zerosFiltered.groupby(group_columns, dropna=False).mean().reset_index()

lab1_df_reduced_zerosFiltered_grouped['diameter'] = lab1_df_reduced_zerosFiltered_grouped['tree_dbh'] + lab1_df_reduced_zerosFiltered_grouped['stump_diam']
print(lab1_df_reduced_zerosFiltered_grouped.head())
lab1_df_reduced_zerosFiltered_grouped.drop(columns=['tree_dbh', 'stump_diam'], inplace=True)
print(lab1_df_reduced_zerosFiltered_grouped.head())
lab1_df_reduced_counts = lab1_df_reduced.assign(count=1)
print(lab1_df_reduced_counts.head())
lab1_df_reduced_counts_grouped = lab1_df_reduced_counts.groupby(group_columns, dropna=False).sum().reset_index()
lab1_df_reduced_counts_grouped.drop(columns=['tree_dbh', 'stump_diam'], inplace=True)
print(lab1_df_reduced_counts_grouped.head())

lab1_df_final = pd.merge(lab1_df_reduced_counts_grouped, lab1_df_reduced_zerosFiltered_grouped, on=group_columns, how='left')
print(lab1_df_final.head())

lab1_df_final.to_csv('/content/lab1_df_final.csv', index=False)
from google.colab import files
files.download('/content/lab1_df_final.csv')
```
