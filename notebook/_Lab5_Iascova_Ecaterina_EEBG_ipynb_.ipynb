```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Titanic dataset
lab2_data = sns.load_dataset('titanic')

# Separate children into boys and girls
lab2_data.loc[(lab2_data['who'] == 'child') & (lab2_data['sex'] == 'male'), 'who'] = 'boy'
lab2_data.loc[(lab2_data['who'] == 'child') & (lab2_data['sex'] == 'female'), 'who'] = 'girl'

# Set custom color palette
palette = {
    'man': '#1f77b4',   # Blue
    'woman': '#ff7f0e', # Orange
    'boy': '#2ca02c',   # Green
    'girl': '#d62728'   # Red
}

# Ensure 'who' column is categorical with specific order
lab2_data['who'] = pd.Categorical(lab2_data['who'], categories=['man', 'woman', 'boy', 'girl'], ordered=True)

# Create the bar plot for survival rate by class and gender/age group
plt.figure(figsize=(10, 6))
sns.barplot(data=lab2_data, x='survived', y='age', hue='who', palette=palette)
plt.title('Survival Rate by Class and Gender/Age Group')
plt.xlabel('Survived')
plt.ylabel('Age')
plt.legend(title='Who')
plt.show()
```

```python
lab2_data.loc[(lab2_data['survived'] == 0), 'survived_str'] = 'no'
lab2_data.loc[(lab2_data['survived'] == 1),  'survived_str'] = 'yes'
lab2_data
```

```python
sns.barplot(data=lab2_data, x='survived', y='age', hue='who')
```
